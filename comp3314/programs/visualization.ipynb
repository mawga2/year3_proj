```python
import pandas as pd
import matplotlib.pyplot as plt
import os
from PIL import Image
```

```python
# Load the training CSV
train_df = pd.read_csv('train.csv')

# Count the number of images per class
label_counts = train_df['label'].value_counts().sort_index()

plt.figure(figsize=(8, 4))
plt.plot(label_counts.index, label_counts.values, marker='o', linestyle='-', color='teal')
plt.title('Class Distribution Overview')
plt.xlabel('Class Label')
plt.ylabel('Number of Images')
plt.xticks(label_counts.index)
plt.grid(alpha=0.3)
plt.legend()
plt.tight_layout()
plt.show()
```

```python
# Parent folder for training images
image_folder = 'train_ims'

# Prepare a figure
plt.figure(figsize=(15, 5))

# Get one image per class
for label in range(10):
    # Get the first image with this label
    img_name = train_df[train_df['label'] == label]['im_name'].values[0]
    img_path = os.path.join(image_folder, img_name)
    
    # Open and plot
    img = Image.open(img_path)
    
    plt.subplot(2, 5, label + 1)
    plt.imshow(img)
    plt.title(f'Class {label}')
    plt.axis('off')

plt.suptitle('One Sample Image from Each Class', fontsize=16)
plt.tight_layout()
plt.show()
```

```python
# Filter the dataset for class 4
class_4_df = train_df[train_df['label'] == 8]

# Prepare a figure to show multiple images
plt.figure(figsize=(15, 10))

# Display 5 random images from class 4
for i, img_name in enumerate(class_4_df['im_name'].sample(5)):
    img_path = os.path.join(image_folder, img_name)
    
    # Open and resize the image
    img = Image.open(img_path)
    img_resized = img.resize((100, 100))  # Resize for better visibility
    
    # Plot the image
    plt.subplot(1, 5, i + 1)
    plt.imshow(img_resized)
    plt.title(f'Class 4')
    plt.axis('off')

plt.suptitle('Sample Images from Class 4', fontsize=16)
plt.tight_layout()
plt.show()
```
